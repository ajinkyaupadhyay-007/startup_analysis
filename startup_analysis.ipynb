```python
import pandas as pd
import numpy as np
df=pd.read_csv('Recently Funded Startups - Sheet1.csv')
df.head()
df.columns=df.columns.str.strip().str.replace(" ","_").str.lower()
def score_funding_round(name):
  if pd.isna(name):
    return 0
  name=name.lower()
  if "series b" in name or "series c" in name:
    return 20
  elif "series a" in name:
    return 15
  elif "seed" in name:
    return 10
  else:
    return 5
df['score_funding_round']=df['round'].apply(score_funding_round)
def score_amount(amount):
    try:
        amount = float(amount)
    except:
        return 0

    if amount >= 100_000_000:
        return 25
    elif amount >= 25_000_000:
        return 15
    elif amount >= 5_000_000:
        return 10
    else:
        return 5
df['score_amount']=df['amount_(usd)'].apply(score_amount)
def score_location(country):
    locations = ["usa", "united states", "uk", "united kingdom", "israel", "germany", "canada"]
    if pd.isna(country):
        return 0
    if country.lower() in locations:
        return 10
    return 5
df['score_location']=df['domain'].apply(score_location)
def score_domain(domain):
    if pd.isna(domain):
        return 0
    fields = ["bio", "pharma", "health", "ai", "med", "drug"]
    domain = domain.lower()
    for kw in fields:
        if kw in domain:
            return 15
    return 5
df['score_domain']=df['domain'].apply(score_domain)
df["probability_score"] = (
    df["score_funding_round"]
    + df["score_amount"]
    + df["score_location"]
    + df["score_domain"]
)
df=df.sort_values(by="probability_score",ascending=False)
df['rank']=np.arange(1,len(df)+1)
final_columns = [
    "rank",
    "company",
    "domain",
    "round",
    "amount_(usd)",
    "country",
    "probability_score"
]
final_df=df[final_columns]
final_df.to_csv('rank_df.csv',index=False)
print("Output saved as ranked_leads.csv")
```
